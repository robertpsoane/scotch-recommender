# Producing example co-occurence graph

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Long, spicy, smoke
# peppery, warmth
# finish, figs, dates, peat, vanilla
```

```python
G = nx.Graph()

G.add_node(1, word="long")
G.add_node(2, word="peppery")
G.add_node(3, word="spicy")
G.add_node(4, word="warmth")
G.add_node(5, word="smoke")
G.add_node(6, word="figs")
G.add_node(7, word="dates")
G.add_node(8, word="peat")
G.add_node(9, word="vanilla")

# Long
G.add_edge(1, 2, weight=1)
G.add_edge(1, 3, weight=2)
G.add_edge(1, 4, weight=1)
G.add_edge(1, 5, weight=2)
G.add_edge(1, 6, weight=1)
G.add_edge(1, 7, weight=1)
G.add_edge(1, 8, weight=1)
G.add_edge(1, 9, weight=1)


# Peppery
G.add_edge(2, 3, weight=1)
G.add_edge(2, 4, weight=1)
G.add_edge(2, 5, weight=1)

# Spicy
G.add_edge(3, 4, weight=1)
G.add_edge(3, 5, weight=2)
G.add_edge(3, 6, weight=1)
G.add_edge(3, 7, weight=1)
G.add_edge(3, 8, weight=1)
G.add_edge(3, 9, weight=1)

# Warmth
G.add_edge(4, 5, weight=1)

# Smoke
G.add_edge(5, 6, weight=1)
G.add_edge(5, 7, weight=1)
G.add_edge(5, 8, weight=1)
G.add_edge(5, 8, weight=1)

# Figs
G.add_edge(6, 7, weight=1)
G.add_edge(6, 8, weight=1)
G.add_edge(6, 9, weight=1)

# dates
G.add_edge(7, 8, weight=1)
G.add_edge(7, 9, weight=1)
```

```python

fig = plt.figure()
fig.patch.set_facecolor('w')

plt.title("Co-occurence graph based on tasting notes from\nLagavulin 16 and Caol Ila 12")
elarge = [(u, v) for (u, v, d) in G.edges(data=True) if d["weight"] == 2]
esmall = [(u, v) for (u, v, d) in G.edges(data=True) if d["weight"] == 1]
pos = nx.spring_layout(G, scale=5)  # positions for all nodes
# nodes
nx.draw_networkx_nodes(G, pos, node_size=700)
nx.draw_networkx_edges(G, pos, edgelist=elarge, edge_color="r", width=3)
nx.draw_networkx_edges(
    G, pos, edgelist=esmall, width=3, alpha=0.5, edge_color="r", style="dashed"
)
# labels
nx.draw_networkx_labels(G, pos, font_size=10, labels={
        1:"long",
        2:"pepp\nery",
        3:"spicy",
        4:"war\nmth",
        5:"smoke",
        6:"figs",
        7:"dates",
        8:"peat",
        9:"vani\nlla"}, font_family="sans-serif")



plt.savefig("Graph.png", dpi=300)
```
