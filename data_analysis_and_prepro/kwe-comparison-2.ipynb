```python
n = 300
txtfile = ""

for key in res:
    output_txt = f"Considering n = {n}, key = {key}\n"
    print(output_txt)
    time_val = res[key]["time"]
    time_str = f"Time = {time_val}\n"
    output_txt = output_txt + time_str
    print(time_str)
    kws = res[key]["kws"]
    if len(kws) > n:
        kws = kws[:n]
    kws = [kw[0] for kw in kws]
    if len(kws) > 20:
        print(kws[:20])
        output_txt = output_txt + "\nTop KWS:\n" + ", ".join(kws[:20])
    else:
        print(kws)
        output_txt = output_txt + "\nTop KWS:\n" + ", ".join(kws)

    vect = ListFeatureVectorizer(kws)
    vectorised = vect.fit(all_lst, True)
    # Adding ID and name to vectorised list
    vectorised["ID"] = n_df.ID
    vectorised["Name"] = n_df.Name
    vectorised_features = vectorised.drop(["ID","Name"], axis=1)
    try:
        # Applying kmeans
        kmeans =  KMeans(
            n_clusters=6,
            n_init=100,
            max_iter=1000
        )

        kmeans.fit(vectorised_features)
        vectorised["kmeans"] = kmeans.labels_

        descs = getClusterDescriptions(kmeans, vectorised_features, 3)
        descs = ";\n ".join([str(k)+": "+", ".join([w[0] for w in descs[k]]) for k in descs])
        print(descs)
        output_txt = output_txt + "Clusters:\n"+ descs + "\n"

        clusters = bottleDetailsFromURLS(vectorised, urls, "kmeans", verbose=False)
        listed = [[k,clusters[k]] for k in clusters]
        listed.sort(key=itemgetter(0))
        clusters = ";\n ".join([str(clust[0]) + ": " + ", ".join(clust[1]) for clust in listed])
        print(clusters)

        output_txt = output_txt +"Clustered Whsikies:\n"+ clusters + "\n"
    except:
        print("No keywords")
        output_txt = output_txt + "No Keywords..."
    txtfile = txtfile + output_txt + "\n\n\n"
        
        
with open("txt-summary.txt", "w") as f:
    f.write(txtfile)       
```
